```python
# 架空の小売業データを作成する

# - store
# - employee
# - customer
# - product
# - shopsales
# - websitelog (TBA)

import random
import re
import itertools
from collections import namedtuple
from datetime import datetime, timedelta

dbname = "TeraShirt"

# define start and end date
start_date = datetime.strptime("2021-05-01", "%Y-%m-%d")
end_date = datetime.strptime("2022-12-31", "%Y-%m-%d")
```

```python
Employee = namedtuple("Employee", "employee_id last_name first_name middle_name gender role report_to salary birthday")

employee_en = """
トム・グリフィン・ブラウン Male デザイナー 13
ジョバンニ・ベルガモ Male デザイナー 13
ブレンダ・クーパー Female 店舗管理 0
"""

employee_ja = """
高橋 大介 Male デザイナー 13
渡辺 拓也 Male 店舗管理 0
小川 博 Male 店舗スタッフ 5
臼井 努 Male 店舗スタッフ 3
高岡 和弘 Male 店舗スタッフ 3
鈴木 あゆみ Female デザイナー 13
長谷川 茜 Female 店舗スタッフ 5
井上 栞 Female 店舗スタッフ 5
望月 千春 Female 店舗スタッフ 3
内藤 沙耶 Female 商品企画 0
梶山 美紀 Female 店舗スタッフ 3
中村 史子 Female 店舗スタッフ 5
"""

def _parse_en(x):
  x = x.split(" ")
  tmp1 = list(reversed(x[0].split("・")))
  if len(tmp1) < 3:
    tmp1.append(None)
  tmp2 = x[1:]
  if len(tmp2) < 2:
    tmp2.append(None)
  else:
    tmp2[-1] = int(tmp2[-1])
  return tmp1 + tmp2
  
def _parse_ja(x):
  x = x.split(" ")
  x.insert(2, None)
  x[-1] = int(x[-1])
  return x

def _salary(role):
  if role.find("管理") >= 0 or role.find("企画"):
    s = random.randint(4500000, 8000000)
  else:
    s = random.randint(3000000, 5000000)
  return int(round(s / 100) * 100)
             
random.seed(87)
employee = list(map(_parse_en, employee_en.strip().split("\n"))) + list(map(_parse_ja, employee_ja.strip().split("\n")))

for e in employee:
  e.append(_salary(e[4]))

for e in employee:
  days = random.randint(365 * 25, 365 * 50)
  bday = end_date - timedelta(days=days)
  e.append(bday.strftime("%Y-%m-%d"))

for i, e in enumerate(employee):
  e.insert(0, i+1)


employee = [Employee(*e) for e in employee]
employee
```

```python
Customer = namedtuple("Customer", "customer_id last_name first_name middle_name last_name_yomi first_name_yomi middle_name_yomi gender birthday register_date")
customers_f = """
吉田 恵	よしだ めぐみ
伊藤 仁美	いとう ひとみ
山本 淳子	やまもと じゅんこ
石井 かおり	いしい かおり
浅野 明子	あさの あきこ
澤田 由美	さわだ ゆみ
秋山 由紀	あきやま ゆき
佐伯 直子	さえき なおこ
片山 早苗	かたやま さなえ
森田 由希子	もりた ゆきこ
園田 明子	そのだ あきこ
小林 朗子	こばやし あきこ
原 麻希	はら まき
成田 奈央	なりた なお
小松 春奈	こまつ はるな
大久保 理紗	おおくぼ りさ
武田 萌子	たけだ もえこ
古川 咲	ふるかわ さき
甲斐 千絵	かい ちえ
遠藤 夕貴	えんどう ゆき
山口 めい	やまぐち めい
中島 亜子	なかじま あこ
山岡 葉子	やまおか ようこ
川本 真由子	かわもと まゆこ
佐竹 美沙	さたけ みさ
平松 あすか	ひらまつ あすか
岩崎 隆子	いわさき たかこ
坂田 美優	さかた みゆ
和田 紫穂	わだ しほ
堀江 小夜子	ほりえ さよこ
丹羽 有里子	にわ ゆりこ
土屋 ユカリ	つちや ゆかり
武本 泰子	たけもと やすこ
大村 冴子	おおむら さえこ
伴 亜季	ばん あき
宮田 早	みやた さき
筒井 礼奈	つつい れな
白石 知江	しらいし ちえ
齋藤 典代	さいとう ふみよ
細山 梢	ほそやま こずえ
今泉 香絵	いまいずみ かえ
桃崎 久美子	ももざき くみこ
吉澤 槙	よしざわ まき
國政 知子	くにまさ ともこ
田辺 利津子	たなべ りつこ
大串 恵利子	おおぐし えりこ
小堺 綾美	こざかい あやみ
高野 聡里	たかの さとり
明石 万由美	あかし まゆみ
豊田 磨希	とよだ まき
小嶋 麻理香	こじま まりか
程 美也子	てい みやこ
安井 麻未	やすい あみ
冨本 智絵	とみもと ちえ
中田 唯那	なかだ ゆいな
鏡山 琴音	かがみやま ことね
早瀬 秋葉	はやせ あきは
福與 瑛里	ふくよ えり
都倉 美夏子	とくら みかこ
栗賀 未知	くりが みち
"""

customers_m = """
中村 太郎	なかむら たろう
吉田 健	よしだ けん
山田 雅彦	やまだ まさひこ
山口 純一	やまぐち じゅんいち
小林 博史	こばやし ひろし
岩崎 学	いわさき まなぶ
山本 雄貴	やまもと ゆうき
藤井 和弘	ふじい かずひろ
齊藤 潤	さいとう じゅん
菊地 翔	きくち しょう
渡邉 裕	わたなべ ゆたか
中村 啓輔	なかむら けいすけ
阿部 正義	あべ まさよし
岡 正人	おか まさと
安藤 遼	あんどう りょう
千葉 篤史	ちば あつし
木村 大志	きむら たいし
原口 健一	はらぐち けんいち
藤井 敬一	ふじい けいいち
今村 祐一	いまむら ゆういち
武内 淳一	たけうち じゅんいち
小林 豊一	こばやし とよかず
大木 隆之	おおき たかゆき
佐藤 恭祐	さとう きょうすけ
河内 誠	こうち まこと
松田 大輝	まつだ ひろき
江口 泰弘	えぐち やすひろ
北村 利夫	きたむら としお
柴田 隼	しばた じゅん
中嶋 敏	なかじま さとし
中嶋 啓	なかじま けい
杉本 宏文	すぎもと ひろふみ
木村 泰紀	きむら やすのり
久保 博久	くぼ ひろひさ
中嶋 隆一	なかしま りゅういち
千葉 岳彦	ちば たけひこ
伊藤 章友	いとう あきとも
大瀧 実	おおたき みのる
安田 寿彦	やすだ としひこ
岡安 啓介	おかやす けいすけ
伊東 秀太	いとう しゅうた
松田 英克	まつだ ひでかつ
笹山 賢治	ささやま けんじ
野村 武敏	のむら たけとし
丹羽 勇作	にわ ゆうさく
白石 英	しらいし すぐる
須田 重之	すだ しげゆき
表 栄一	おもて えいいち
久保 裕成	くぼ ひろなり
岩崎 慎司	いわざき しんじ
土橋 勝哉	どばし かつや
磯崎 誠人	いそざき まさと
窪園 進	くぼぞの すすむ
岩土 翔	いわど しょう
中津留 大志	なかつる たいし
今堀 裕晃	いまほり ひろあき
一寸木 琢也	ちょっき たくや
冨谷 航大	とみや こうだい
"""

def _parse_customer(x, gender):
  x = re.sub("\s", " ", x.strip()).split(" ")
  x.insert(2, None)  # missing middle name
  x.insert(4, None)  # missing middle name yomi
  x.append(gender)
  if random.random() < 0.1:  # missing rate
    x.append(None)
  else:
    days = random.randint(365 * 25, 365 * 50)
    bday = end_date - timedelta(days=days)
    x.append(bday.strftime("%Y-%m-%d"))  

  if random.random() < 0.2:
    # registered at the open day
    register = start_date
  else:
    register = start_date + timedelta(days=random.randint(0, 600))
  x.append(register.strftime("%Y-%m-%d"))
  return x

random.seed(87)
customers = [_parse_customer(c, "Female") for c in customers_f.strip().split("\n")] + [_parse_customer(c, "Male") for c in customers_m.strip().split("\n")]
for i, c in enumerate(customers):
  c.insert(0, i+1)
customers = [Customer(*c) for c in customers]
customers
```

```python
Store = namedtuple("Store", "store_id name name_en region prefecture open_date store_manager")
stores = [
  ["３番通り店", "3rd Avenue", "関東", "東京都", 5]
  ,["中央駅前店", "Central Station", "関東", "神奈川県", 5]
  ,["空港第２ターミナル店", "千葉県", "関東", "千葉県", 3]
  ,["赤坂山アウトレット店", "東北", "厚本市", "宮城県", 3]
]
def _open_date(start_date, daydiff):
  d = start_date + timedelta(days=daydiff)
  return d.strftime("%Y-%m-%d")


opendates = [
  _open_date(start_date, 0),
  _open_date(start_date, 7*25),
  _open_date(start_date, 7*40),
  _open_date(start_date, 7*70)
]
                    

for i, s in enumerate(stores):
  s.insert(4, opendates[i])
  s.insert(0, i+1)
stores = [Store(*s) for s in stores]
stores
```

```python
Product = namedtuple("Product", "product_id name type color size target standard_unit_price")

colors = ["orange", "white", "black", "gray"]
sizes = ["XS", "S", "M", "L", "XL"]
types = ["tshirt", "longt", "hoody"]
types_ja = dict(zip(types, ["Tシャツ", "ロングT", "パーカー"]))
targets = ["Men", "Women", "Kids"]

def _unit_price(x):
  t = x[0]
  price = (
    1000 if t == "tshirt" else
    1500 if t == "longt" else
    2500 if t == "sweater" else
    3000 if t == "hoody" else
    None
  )
  if x[3] == "Women":
    price *= 1.5
  elif x[3] == "Men":
    price *= 2

  if x[2] == "XS":
    price *= 0.8
  elif x[2] == "S":
    price *= 0.9
  elif x[3] == "L":
    price *= 1.1
  elif x[2] == "XL":
    price *= 1.5  
  price = int(round(price / 50) * 50)
  return price

def _name(x):
  return "{} ({}, {}, {})".format(types_ja[x[0]], x[3], x[2], x[1])

products = list(itertools.product(types, colors, sizes, targets))
products = list(list(p) for p in products)
for i, p in enumerate(products):
  p.append(_unit_price(p))
  p.insert(0, _name(p))
  p.insert(0, i+1)
products = [Product(*p) for p in products]
products
```

```python
StoreSales = namedtuple("StoreSales", "purchase_date purchase_timestamp store_id customer_id cashier_id product_id unit_price discount quantity sales_value")

def _discount_price(date, product):
  price = product.standard_unit_price
  md = date.month, date.day
  if md >= (12, 10) or md < (1, 10):
    price *= 0.7
  if date.month in (9, 10) and product.type == "tshirt":
    price *= 0.9
  if date.month in (3, 4, 5) and product.type == "hoody":
    price *= 0.95
  return round(price)

def _get_timestamp(date):
  hour = random.choices(list(range(10, 21)), [1,1,1,3,4,4,3,2,1,1,1])[0]
  second = random.randint(0, 59)
  out = date + timedelta(hours=hour, seconds=second)
  return out

def _get_cashier(store):
  candidates = [store.store_manager] + [e.employee_id for e in employee if e.report_to == store.store_manager]
  weights = [1] * len(candidates)
  weights[0] = 0.2
  return random.choices(candidates, weights)[0]
  
def _get_product_weight(customer, date, base_product_weight):
  if customer is None:
    return base_product_weight.copy()

  weight = base_product_weight.copy()
  def _adjust_weight(p, w):
    if customer is not None:
      if customer.gender == "Female":
        if p.target == "Women":
          w *= 1.5
        elif p.target == "Men":
          w *= 0.3
      elif customer.gender == "Men":
        if p.target == "Women":
          w *= 0.5
        elif p.target == "Men":
          w *= 1.3
    if date.month >= 6 and date.month <= 9:
      if p.type == "hoody":
        w *= 0.1
    return w
  weight = [_adjust_weight(p, w) for p, w in zip(products, weight)]

  return weight

def _get_store_weight(base_store_weight, date):
  weight = [0 if s.open_date > date.strftime("%Y-%m-%d") else w for s, w in zip(stores, base_store_weight)]  
  return weight

def _person_visit(customer, date, base_product_weight, base_store_weight, prob_more):
  store_weight = _get_store_weight(base_store_weight, date)
  store = random.choices(stores, store_weight)[0]
  cashier = _get_cashier(store)
  
  timestamp = _get_timestamp(date)

  customer_id = None if customer is None else customer.customer_id
  product_weight = _get_product_weight(customer, date, base_product_weight)
  out = []
  while True:
    product = random.choices(products, product_weight)[0]
    unit_price = _discount_price(date, product)
    discount = product.standard_unit_price - unit_price
    quantity = round(random.expovariate(3)) + 1
    s = [date.strftime("%Y-%m-%d"), timestamp.strftime("%Y-%m-%d %H:%M:%S+09:00"), store.store_id, customer_id, cashier, product.product_id, unit_price, discount, quantity, quantity*unit_price]
    out.append(s)
    if random.random() > prob_more:
      break
  return out

def _person_history(customer):
  # probability to purchase one more item
  prob_more = random.random() * 0.33
  # set preference for products
  base_product_weight = [random.random() for p in products]
  
  # set average days to next visit
  mu_day_next = random.random() * 90 + 30  # 30 to 500 days
  # set preference for a store
  base_store_weight = [random.random() for s in stores]
  
  out = []
  if customer is None:
    day = start_date
  else:
    day = datetime.strptime(customer.register_date, "%Y-%m-%d")
  while day <= end_date:
    out += _person_visit(customer, day, base_product_weight, base_store_weight, prob_more)    
    daydiff = random.expovariate(1.0 / mu_day_next)
    day = day + timedelta(days=round(1 + daydiff))

  return out

random.seed(87)
store_sales = []
for c in customers:
  store_sales += _person_history(c)
for _ in range(150):
  store_sales += _person_history(None)
  
store_sales = [StoreSales(*s) for s in store_sales]
store_sales
```
