```python
import pandas as pd
from urllib.request import urlopen
u = urlopen("https://www8.cao.go.jp/chosei/shukujitsu/syukujitsu.csv")
x = pd.read_csv(u, encoding="cp932")
x.columns = ["calendar_date", "holiday_name"]
x.calendar_date = pd.to_datetime(x.calendar_date, format="%Y/%m/%d").dt.strftime("%Y-%m-%d")
x
```

```python
import os

savepath = "../data/japan/holidays/data.csv"
os.makedirs(os.path.dirname(savepath), exist_ok=True)
x.to_csv(savepath, index=False, encoding="utf8")
```
